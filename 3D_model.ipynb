```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
xs =[1,70,138,207,275,343,412,481,548]
ys =[1,5311,332424,3044940,8088851,10374932,11409831,26752447,31528114]
zs =[0,1.91,1.2,1.08,0.89,0.95,1.34,0.79,1.01]

ax.scatter(xs, ys, zs, c='r', marker='^')

ax.set_xlabel('Days')
ax.set_ylabel('Total Cases in Millions')
ax.set_zlabel('Reproduction Rate')

plt.show()
```
